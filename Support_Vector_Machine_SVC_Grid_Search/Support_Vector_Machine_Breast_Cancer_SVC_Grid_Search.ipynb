```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
%matplotlib inline
```

### Importing data

```python
data = load_breast_cancer()
```

```python
data.keys()
```

### Creating a data frame

```python
df_feat = pd.DataFrame(data['data'],columns=data['feature_names'])
df_feat
```

```python
df_feat.describe()
```

### Train, test, split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_feat
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### SVM machine learn

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

### Getting parameters

```python
model.get_params()
```

### Predicting

```python
pred = model.predict(X_test)
```

### Creating a class report and confusion matrix

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

### Searching for the best parameters with grid search 

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1,10,100,1000],'gamma': [1,0.1,0.01,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(),param_grid,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

### Getting the best parameter from grid search 

```python
grid.best_params_
```

### Predicting with grid search

```python
grid_pred = grid.predict(X_test)
```

### Class report and confusion matrix

```python
print(classification_report(y_test,grid_pred))
print('\n')
print(confusion_matrix(y_test,grid_pred))
```
