```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

### Creating a columns variable

```python
columns_names = ['user_id','item_id','rating','timestamp']
```

### Importing data

```python
df = pd.read_csv('u.data',sep='\t',names=columns_names)
df
```

### Getting the movie titles

```python
mt = pd.read_csv('Movie_Id_Titles')
mt
```

### Merging data frames

```python
df = pd.merge(df,mt,on='item_id')
df
```

### Creating a data frame with avg rating and number of rating

```python
df_rate_avg = pd.DataFrame(df.groupby('title')['rating'].mean().sort_values(ascending=False))
df_rate_avg
```

```python
df_rate_numb = pd.DataFrame(df.groupby('title')['rating'].count().sort_values(ascending=False))
df_rate_numb
```

### Merging data frames

```python
ratings = pd.merge(df_rate_avg,df_rate_numb,on='title',suffixes = ('_avg', '_num'))
ratings
```

### Creating histograms using the new data frame

```python
ratings['rating_num'].hist(bins=70)
```

```python
ratings['rating_avg'].hist(bins=70)
```

### Analizing the relationship between rate_avg and rate_num

we can visualize that there is a trend to a higher number of ratings in movies with a higher rating score 

```python
sns.jointplot(x='rating_avg',y='rating_num',data=ratings)
```

### Creating a matrix that has a user_id in one axis and the movie titles in another

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat
```

```python
ratings.sort_values('rating_num',ascending=False).head(10)
```

```python
starwars_rating = moviemat['Star Wars (1977)']
liarliar_rating = moviemat['Liar Liar (1997)']
```

```python
starwars_rating
```

### Getting the correlations between series 

```python
sim_starwars = moviemat.corrwith(starwars_rating)
sim_starwars
```

```python
sim_liarliar = moviemat.corrwith(liarliar_rating)
```

### Cleaning the null values

```python
corr_starwars = pd.DataFrame(sim_starwars,columns=['correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars
```

### Trying to sorting by correlation to get the most similar movies

It does not work because there are a many movies with a low quantity of ratings creating noise in the analysis 

```python
corr_starwars.sort_values('correlation',ascending=False)
```

### Removing movies with low rating number

Analysing the histogram I concluded that the best way to remove the noise is dropping movies with less than 100 rates

```python
ratings['rating_num'].hist(bins=70)
```

```python
corr_starwars = corr_starwars.join(ratings['rating_num'])
```

### Getting a high correlated movie

```python
corr_starwars[corr_starwars['rating_num']>100].sort_values('correlation',ascending=False).head(20)
```

### Trying to create the same correlation with another movie

```python
corr_liarliar = pd.DataFrame(sim_liarliar,columns=['correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['rating_num'])
corr_liarliar
```

### Getting a high correlated movie

```python
corr_liarliar[corr_liarliar['rating_num']>100].sort_values('correlation',ascending=False).head(20)
```
